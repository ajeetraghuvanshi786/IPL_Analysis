```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("matches.csv")
```

```python
df.head(2)
```

```python
df = df.rename(columns={'team1':'fielding','team2':'bating'})
```

```python
df.head(2)
```

```python
result = df.groupby(['season','winner']).size().reset_index(name = 'winning_count')
result = result.pivot(index='season', columns='winner', values='winning_count').fillna(0).astype(int)
```

```python
result
```

```python
df['fielding'].unique()
```

```python
df['batting'].unique().tolist()
```

```python
result['Total Matches'] = result.sum(axis=1)
```

```python
result
```

```python
# Winner Teams of Each Season

result = df[df['match_type'] == 'Final']
```

```python
result
```

```python
result = result[['season','winner','venue','toss_winner','toss_decision']].reset_index(drop=True)
result = result.rename({'season':'Season','winner':'Winner'})
```
